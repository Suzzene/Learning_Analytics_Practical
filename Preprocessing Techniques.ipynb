```python
# Q2: Data Pre-processing Techniques using Pandas
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('student_data.csv')
```

```python
print("Initial Dataset Info:")
print(data.info())
```

```python
data.columns = data.columns.str.strip().str.replace('"', '').str.replace('\t', '')
```

```python
print("\nCleaned Column Names:")
print(data.columns.tolist())
```

```python
print("\nMissing Values in Each Column:")
print(data.isnull().sum())
```

```python
num_cols = data.select_dtypes(include=[np.number]).columns
data[num_cols] = data[num_cols].fillna(data[num_cols].mean())

cat_cols = data.select_dtypes(include=['object']).columns
for col in cat_cols:
    data[col] = data[col].fillna(data[col].mode()[0])
```

```python
data.drop_duplicates(inplace=True)
```

```python
if 'Age at enrollment' in data.columns:
    data['Age at enrollment'] = data['Age at enrollment'].astype(int)
```

```python
data_encoded = pd.get_dummies(data, drop_first=True)
```

```python
selected_data = data_encoded.iloc[:, :10]
```

```python
print("\nPreprocessed Dataset (First 5 Rows):")
print(selected_data.head())

print("\nFinal Dataset Shape:", selected_data.shape)
```
