```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('student_data.csv')
```

```python
data.columns = data.columns.str.strip().str.replace('"', '').str.replace('\t', '')
```

```python
num_data = data.select_dtypes(include=[np.number])
```

```python
# 1. Line Plot
plt.figure()
plt.plot(num_data.iloc[:, 0], num_data.iloc[:, 1])
plt.title("Line Plot")
plt.xlabel(num_data.columns[0])
plt.ylabel(num_data.columns[1])
plt.show()
```

```python
# 2. Bar Chart
plt.figure()
plt.bar(num_data.iloc[:, 0], num_data.iloc[:, 1])
plt.title("Bar Chart")
plt.xlabel(num_data.columns[0])
plt.ylabel(num_data.columns[1])
plt.show()
```

```python
#Histogram
plt.figure()
plt.hist(num_data.iloc[:, 0], bins=10)
plt.title("Histogram")
plt.xlabel(num_data.columns[0])
plt.ylabel("Frequency")
plt.show()
```

```python
#Scatter Plot
plt.figure()
plt.scatter(num_data.iloc[:, 0], num_data.iloc[:, 1])
plt.title("Scatter Plot")
plt.xlabel(num_data.columns[0])
plt.ylabel(num_data.columns[1])
plt.show()
```

```python
#Box Plot
plt.figure()
sns.boxplot(y=num_data.iloc[:, 0])
plt.title("Box Plot")
plt.show()
```
